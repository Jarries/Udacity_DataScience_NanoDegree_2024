### Some EDA on the combined dataframe

```python
def eda_function(df):
    """Input: A dataframe contains multiple columns
    
        Output: All the unique values of each column, together with the count of nulls in each column.
    """
    for column in df.columns:
        if column != 'year_of_survey':
            print("Column Name: " + column + '\n')
            print("Unique Values: \n" + str(df[column].unique()))
            print("Null values: " + str(df[column].isnull().sum()))
            print("Null percentage: " + str(df[column].isnull().sum()/len(df) * 100) + ' %')
    
            # Group by 'Category' and count the amount of nulls in each group
            null_counts = df.groupby('year_of_survey').apply(lambda x: x.isnull().sum())
            null_counts = null_counts.drop(['year_of_survey'], axis=1)
    
            # Calculate the total number of rows in each group
            total_counts = df.groupby('year_of_survey').size().reset_index(name='Total')
    
            # Merge the null counts with the total counts
            null_counts = null_counts.reset_index()
            
            null_counts = pd.merge(null_counts, total_counts, on='year_of_survey')
    
            # Calculate the percentage of nulls
            null_counts['null_percentage'] = (null_counts[column] / null_counts['Total']) * 100
    
            display(null_counts[['year_of_survey', 'null_percentage']])
            print("\n")
```
